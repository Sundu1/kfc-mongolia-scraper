```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
path = r'C:\Users\user\Desktop\projects\webscrap\chromedriver.exe'
# path = r'C:\Program Files (x86)\Google\Chrome\Application\chrome.exe'
driver = webdriver.Chrome(path)
driver.get("https://www.vocabulary.com/lists/191545")

length = driver.find_elements(By.CLASS_NAME, 'word')
words = []
for i in range(0,len(length)):
    try:
        word = driver.find_elements(By.CLASS_NAME, 'word')[i].text
        words.append(word)
    except IndexError:
        print("it's out of index, son")
```
